```python
dobre_haslo='buziaczek12.pl'
tajna_wiadomość='nigdy nie jedz żółtego śniegu!'

user_mowi=input('podaj hasło')
if user_mowi==dobre_haslo:
    print('to jest tajna wiadomość')
    print(tajna_wiadomość)
else:
    print('złe hasło')
```

```python
dobre_haslo='buziaczek12.pl'
tajna_wiadomosc='nigdy nie jedz żółtego śniegu!'

user_mowi=input('podaj hasło')

while user_mowi!=dobre_haslo:
    print('złe haslo')
    user_mowi=input('podaj hasło')
print('to jest tajna wiadomość')
print(tajna_wiadomosc)
```
